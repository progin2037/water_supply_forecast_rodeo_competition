# July (month 7)

```python
jul_path = 'results/hyperparams_tuning/study_initial_2024_03_24_20_37_28_month_7.pkl'
jul_results = joblib.load(jul_path).trials_dataframe()
```

```python
jul_time = jul_results['datetime_complete'].iloc[-1] - jul_results['datetime_start'].iloc[0]
print(jul_time)
time_to_execute += jul_time
```

```python
jul_results.sort_values('value').head(10)
```

```python
jul_results.sort_values('user_attrs_best_result_without_rms').head(10)
```

## Conclusions
* results
    * model with best RMS value (28.347714) is only 10th for values without RMS
    * best value without RMS is different, not even from TOP 10 with RMS (22.676767 without and 28.905824 with RMS)
    * <font color = 'red'>Too much interval coverage (0.852273). Use less DISTR_PERC for the final tuning, change it from 0.1 to 0.05</font>
* features
    * models were expected to get better performance. Because of that, some features were appended/deleted. Features deleted for the final tuning are crossed out, features added to the final tuning are bolded:
        * site_id
        * nat_flow_prev
        * issue_date_no_year
        * longitude
        * nat_flow_Apr_mean
        * WTEQ_DAILY_Apr_mean
        * pdsi_prev
        * <s>pdsi_prev_to_last_month_diff</s>
        * <b>sd_prev</b>
        * <b>WTEQ_DAILY_Jul_prev_diff</b>
* hyperparams summary (to determine values range for the final tuning)
    * feature_fraction 0.875 works best. There is also one in top 10 with feature fraction 1.0, so it is worth to keep
    * lambda_l2 takes various values - from 0.02 to 1.6. These values however aren't too big for best models. Include also a little higher values in final tuning (less overfitting)
    * best results are for learning rate between 0.02 - 0.08
    * max_bin varies heavily (150-300...). Don't use higher values, so training doesn't take too long
    * best max_depth is between 8-10 (usually 10). Don't use higher values to not overfit
    * min_data_in_leaf is between 15-25. Use a little higher to not overfit
    * num_leaves between 57-122. Use higher values to speed up training
    * subsample between 0.76-0.88 with 0.93 outlier
